```python
df = pd.read_csv('../data/wine.csv', index_col=0)
df
```

```python
df.points.describe()
```

```python
df.points.unique()
```

```python
df.taster_name.value_counts()
```

```python
# map

mean = df.points.mean()
df.points.map(lambda x : x - mean)
```

```python
# apply is applied to while dataframe

def remean_points(row) : 
    row.points = row.points - mean
    return row

df.apply(remean_points, axis='columns')
```

```python
df.country + "-" + df.region_1
```

```python
reviews = df
highest_points_to_price = (reviews.points / reviews.price)
maxi = highest_points_to_price.max()
bargain_wine = reviews[highest_points_to_price == maxi]
bargain_wine
```

```python
bargain_wine.reset_index(drop=True, inplace= True)
bargain_title = bargain_wine.loc[0, 'title']
bargain_title
```

```python
tropical_count = reviews['description'].map( lambda s : "tropical" in s  ).sum()
fruity_count = reviews['description'].map(lambda s : "fruity" in s).sum()
descriptor_counts = pd.Series([tropical_count, fruity_count])
descriptor_counts
```

```python


def reduce(row) : 
    
    
    
    if row.points >= 95 or row.country == 'Canada' : 
        return 3
    elif row.points >= 85 and row.points <95  : 
        return 2   
    
    return 1

star_ratings = reviews.apply(reduce,axis='columns')
star_ratings
```
